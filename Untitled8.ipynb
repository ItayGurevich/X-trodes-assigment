```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt
```

```python
class SignalProcessor:
    def __init__(self, file_path, num_channels, sample_rate, voltage_resolution, num_ADC_bits):
        self.file_path = file_path
        self.num_channels = num_channels
        self.sample_rate = sample_rate
        self.voltage_resolution = voltage_resolution
        self.num_ADC_bits = num_ADC_bits
        self.data = self.load_data()
        self.df = self.data_to_dataframe()

    def load_data(self):
        # Loads the data from the binary file and converts it to voltage values
        data = np.fromfile(self.file_path, dtype=np.uint16)
        data = np.reshape(data, (self.num_channels, -1), order='F')
        data = np.multiply(self.voltage_resolution, (data - np.float_power(2, self.num_ADC_bits - 1)))
        return data

    def data_to_dataframe(self):
        # Converts the data to a Pandas DataFrame
        df = pd.DataFrame(self.data.T, columns=[f'Channel_{i}' for i in range(1, self.num_channels + 1)])
        return df

    def print_signals(self, start=0, end=10000):
        # Displays the original signals over time
        fig, axes = plt.subplots(self.num_channels, 1, figsize=(15, 3 * self.num_channels))
        time = np.arange(start, end) / self.sample_rate
        for i, column in enumerate(self.df.columns):
            axes[i].plot(time, self.df[column][start:end])
            axes[i].set_title(f'Signal - {column}')
            axes[i].set_xlabel('Time (seconds)')
            axes[i].set_ylabel('Voltage (V)')
        plt.tight_layout()
        plt.show()

    def bandpass_filter(self, lowcut, highcut):
        #  Applies a band-pass filter according to the desired frequency range
        nyquist = 0.5 * self.sample_rate
        low = lowcut / nyquist
        high = highcut / nyquist
        b, a = butter(4, [low, high], btype='band')
        for column in self.df.columns:
            self.df[column] = filtfilt(b, a, self.df[column])

    def plot_signals(self, start=0, end=10000):
        # Displays the filtered signals over time
        fig, axes = plt.subplots(self.num_channels, 1, figsize=(15, 3 * self.num_channels))
        time = np.arange(start, end) / self.sample_rate
        for i, column in enumerate(self.df.columns):
            axes[i].plot(time, self.df[column][start:end])
            axes[i].set_title(f'Filtered Signal - {column}')
            axes[i].set_xlabel('Time (seconds)')
            axes[i].set_ylabel('Voltage (V)')
        plt.tight_layout()
        plt.show()

    def plot_frequency_spectrum(self):
        # Calculates and displays the frequency spectrum for each channel in decibels
        fig, axes = plt.subplots(self.num_channels, 1, figsize=(15, 3 * self.num_channels))
        for i, column in enumerate(self.df.columns):
            signal = self.df[column]
            freqs = np.fft.fftfreq(len(signal), d=1/self.sample_rate)
            fft_vals = np.fft.fft(signal)
            fft_vals_db = 20 * np.log10(np.abs(fft_vals) + 1e-6)
            axes[i].plot(freqs[:len(freqs)//2], fft_vals_db[:len(freqs)//2])
            axes[i].set_title(f'Frequency Spectrum - {column}')
            axes[i].set_xlabel('Frequency (Hz)')
            axes[i].set_ylabel('Amplitude (dB)')
        plt.tight_layout()
        plt.show()
```

```python

file_path = r"/content/NEUR0000 (1).DT8"
num_channels = 8
sample_rate = 4000
voltage_resolution = 4.12e-7
num_ADC_bits = 15

processor = SignalProcessor(file_path, num_channels, sample_rate, voltage_resolution, num_ADC_bits)
processor.plot_frequency_spectrum()
processor.print_signals(start=400, end=1000)
processor.bandpass_filter(300, 350)
processor.plot_signals(start=400, end=1000)
```
